```python
import matplotlib.pyplot as pp
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
x = np.linspace(0,5,11)
y = np.sin(x)
y
```

```python
fig = pp.figure(figsize=(10,3))
axes = fig.add_axes([0.1,0.1,0.8,0.8]) #left, bottom, width, height (range 0 to 1)
axes.plot(x,y,'r')
```

```python
# Creating Figure and axis using subplots method
# we can specify number of rows and columns in the subplots() in which case we get a tuple in the axes return value.
fig, axes = pp.subplots(figsize=(6,4)) 
axes.plot(x, y, 'b')
axes.set_xlabel('Set X Label') # can set the labels, limits ,scale this way as well.
axes.set_ylabel('Set Y Label')
axes.set_title('Set Title')
pp.show()
```

```python
'''Nested Plots'''
fig = pp.figure(figsize=(10,4))
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.65,0.5,0.2,0.3])

# Larger figure
axes1.plot(x, y, 'b')
axes1.set_xlabel('Set X Label')
axes1.set_ylabel('Set Y Label')
axes1.set_title('Set Outer Title')

# Inner figure
axes2.plot(x, -y, 'r')
axes2.set_xlabel('Inner X Label')
axes2.set_ylabel('Inner Y Label')
axes2.set_title('Set Inner Title')
```

```python
'''Multiple plots'''
fig, axes = pp.subplots(nrows=1, ncols=2, figsize = (10,4))
#axes in this case is an array
(axes1, axes2) =axes #unpack the tuple

axes1.plot(x, y, 'b')
axes1.set_xlabel('Set X1 Label')
axes1.set_ylabel('Set Y1 Label')
axes1.set_title('Set Title 1')

axes2.plot(x, -y, 'r')
axes2.set_xlabel('Inner X2 Label')
axes2.set_ylabel('Inner Y2 Label')
axes2.set_title('Set Title 2')

pp.tight_layout()# use this to adjust spacing so that there is no overlapping of content 
```

```python
''' Legends Limits, Scale, Ticks, Labels'''
fig = pp.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x, y, label='Sinx')
axes.plot(x, np.cos(x), label='Cosx')
```

```python
fig = pp.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x, y, label='Sinx')
axes.plot(x, np.cos(x), label='Cosx')
axes.set_xscale('log')
axes.legend(title ='My Legend', loc='best')
```

```python
fig = pp.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x, y, label='Sinx')
axes.plot(x, np.cos(x), label='Cosx')
axes.set_xlim(-10, 10)
axes.set_ylim(-10,10)
```

```python
''' Line width, style,  color'''
fig = pp.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x, y, label='Sinx', lw = 2, ls = ':', c = 'r')
axes.plot(x, np.cos(x), label='Cosx', lw = 3, ls = '--', c = 'g')
```

```python
'''Grid'''
fig = pp.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x, y, label='Sinx', lw = 2, ls = ':', c = 'r')
axes.plot(x, np.cos(x), label='Cosx', lw = 3, ls = '--', c = 'g')
axes.set_facecolor('lightgrey')
```
