```python
import numpy as np
import pandas as pd
```

```python
football = pd.read_csv('football.csv')
football
```

```python
# where wins greater than 10
football[football['wins'] > 10] # NOTE the wins columns with Nan is ignored
```

```python
# selecting specific columns conditionally
football[football['wins'] > 10][['game_id', 'year','team', 'wins']]
```

```python
# OR recommended way if the selection is meant for assignment.
football.loc[football['wins'] > 10, ['game_id', 'year', 'team', 'wins']]
```

```python
# MULTIPLE Conditions
# Note the brackets are required for seperating the conditions, Also single &, | are used
football[ (football['wins'] < 10) & (football['losses'] < 5) ] 
```

```python
''' Conditional Assignments '''

# make all teams with wins more than 10, losses as 0 
football[football['wins'] > 10]['losses'] = 0

# The error below indicates that we are trying to assign value to a dataframe which is the copy of our dataframe.
# in the above example the first [] will return a different dataframe  containing records where wins > 10
# and then the losses column from this copy will be assigned the value 0 which is not what we intended.
```

```python
# the Loc will get the records which satisfies the condition (as this specified in the row_indexer position)
# and the losses column in one shot. Additionally it returns the dataframe itself and not the copy.
football.loc[football['wins'] > 10, 'losses'] = 0 
football
```
